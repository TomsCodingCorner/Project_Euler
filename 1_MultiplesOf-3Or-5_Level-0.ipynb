```python
def multiplication_3_or_5(below=1000)-> int:
    list = []

    for i in range(1, below):
        if i/3 % 1 == 0 or i/5 % 1 == 0:
            list.append(i)
    
    return sum(list)

print(multiplication_3_or_5())
```
