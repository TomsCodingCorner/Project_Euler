```python
def gen_sequence(limit = 4000000):
    sequence = [0, 1]
    while sequence[-1] < limit:
        sequence.append(sequence[-2]+sequence[-1])
    newlist = []
    for i in sequence[:-1]:
        if i % 2 == 0:
            newlist.append(i)
    return sum(newlist)
print(gen_sequence())
```
