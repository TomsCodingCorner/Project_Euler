```python
def sum_primes_below(number = 2000000)-> int:
    """
    Function counting the primes bellow number
    """
    if number < 2:
        return 0
    
    total = 2 #Waarde 2 doen we we buiten de loop
    for num in range(3, number+1, 2):
        for check in range(3, int(num ** 0.5) + 1,2):
            if num % check == 0:
                break
        else:
            total += num
    return total

print(sum_primes_below())
```
