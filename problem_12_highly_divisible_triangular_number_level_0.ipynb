```python
def triangular_dibidible_by(goal_dividors: int)->int:
    num = 1
    triangle = 1
    while(True):
        count = 0
        for check in range(1, int(triangle ** 0.5)+1):
            if triangle % check == 0:
                count += 2
        if int(triangle ** 0.5) ** 2 == triangle:
            count -= 1
        if count > goal_dividors:
            return triangle
        num += 1
        triangle += num

print(triangular_dibidible_by(500))
```
