```python
def smallest_multiple()-> int:
    current = 2520
    while True:  
        for divisibible in [16, 9, 5, 7, 11, 13, 17, 19]:
            if current % divisibible != 0:
                break
        else:
            return current
        current += 20
            
print(smallest_multiple())
```
