```python

def calc_down_to_1(n):
    count = 0
    while(int(n)!=1):
        if n % 2 == 0: #even
            n = n/2
            count += 1
        else: #uneven
            n = 3*n + 1
            count += 1
    return count

def longest_collatz_sequenze(n=1000000)-> int:
    current_best = 0
    best_val = 0
    for i in range(1, n):
        tries = calc_down_to_1(i)
        if tries > current_best:
            best_val = i
            current_best = tries
    return best_val
   
print(longest_collatz_sequenze())
```
