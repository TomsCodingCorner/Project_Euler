```python
def letter_to_length(num) -> int:
    """
    Return the number of letters in the written form of a number.
    e.g: 124 -> one hundred and twenty four = 24
    """
    length = 0
    one_to_nineteen = [3,3,5,4,4,3,5,5,4,3,6,6,8,8,7,7,9,8,8]
    twenty_to_90 = [6,6,5,5,5,7,6,6]

    last_one = num%10
    last_two = num%100
    hundreds = (num // 100) % 10

    if last_two != 0:
        if last_two <= 19:
            length += one_to_nineteen[last_two-1]
        else:
            length += twenty_to_90[(last_two // 10) - 2]
            if last_one != 0:
                length += one_to_nineteen[last_one-1]
    
    if num >= 100 and hundreds != 0:
        length += 7 #Account for the "Hundred" (100)
        length += one_to_nineteen[hundreds-1]
        if last_two != 0:
            length += 3 #Account for "And"

    if num == 1000: #For the last one thousand
        return 11    
    return length
```

```python
count_total_length = 0
for i in range(1000):
    count_total_length += letter_to_length(num=i+1)

print(count_total_length)
```
