```python


def letter_to_length(num) -> int:
    length = 0
    one_to_12 = [3,3,5,4,4,3,5,5,4,3,6,6]
    twenty_to_90 = [6,6,5,5,5,7,6,6]

    num_length = len(str(num))
    example = 100

    if str(num)[-2:] != str(example)[-2:]:
        if int(str(num)[-2:]) < 13 and str(num)[-2:] != str(00):
            length += one_to_12[int(str(num)[-2:])-1]
        else:
            length += twenty_to_90[int(str(num)[-2])-2]
            length += one_to_12[int(str(num)[-1])-1]
    
    if num_length >= 3 and str(num)[-3] != str(0):
        #print("Passing 2")
        length += 3 #Account for "And"
        length += 7 #Account for the "Hundrerd" (100)
        length += one_to_12[int(str(num)[-3])-1]


    if num_length == 4: #For the lost one thousand
        #print("Passing 3")
        length += 11 
    
    return length
```

```python
count_total_length = 0
for i in range(1000):
    count_total_length += letter_to_length(num=+1)

print(count_total_length)
```

```python
#Ground truths
print(letter_to_length(342)) # 23
print(letter_to_length(115)) # 20
```

```python
print(letter_to_length(1)) # 3
print(letter_to_length(2)) # 3
print(letter_to_length(3)) # 5
print(letter_to_length(12), "\n") # 6

print(letter_to_length(32)) # 9
print(letter_to_length(44)) # 10
print(letter_to_length(88), "\n") # 11

print(letter_to_length(111)) # 20
print(letter_to_length(44)) # 10
print(letter_to_length(88)) # 11
print(letter_to_length(1000), "\n") # 11
```
