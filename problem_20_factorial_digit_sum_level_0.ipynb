```python

def sum_of_digits(num):
    """
    Ez, probably can be shorter in code and faster, but readability is something too :)
    """
    count = 1
    for i in range(num):
        count *= i+1
    digits = 0
    for i in range(len(str(count))):
        digits += int(str(count)[i])
    return digits

sum_of_digits(100)
    
```
