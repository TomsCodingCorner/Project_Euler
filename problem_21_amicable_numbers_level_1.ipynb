```python
def proper_division_sum(n):
    total = 0
    for i in range(1,n):
        if n % i == 0:
            total += i
    return total

def main(limit):
    total = 0
    for a in range(1, limit):
        b = proper_division_sum(a)
        if b != a and proper_division_sum(b) == a:
            total += a
    return total

main(10000)
```
