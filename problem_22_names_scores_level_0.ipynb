Reading in the data

```python
data = open("data/0022_names.txt")
data = data.read()
data = data.replace('"', '')
data = data.split(",")
data.sort()
```

Main code

```python
def idx_in_alfabet(letter: str) -> int:
    alfabet = "abcdefghijklmnopqrstuvwxyz"
    for letter_idx in range(len(alfabet)):
        if alfabet[letter_idx] == letter or alfabet[letter_idx].upper() == letter:
            return letter_idx+1
    return -1

def main(name_list: list) -> int:
    total_score = 0
    for name_idx in range(len(name_list)):
        name_score = 0
        for letter in name_list[name_idx]:
            name_score += idx_in_alfabet(letter)
        total_score += name_score * (name_idx+1)
    return total_score
print(main(name_list = data))
```
