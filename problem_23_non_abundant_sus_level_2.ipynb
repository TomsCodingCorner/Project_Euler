```python
def is_abundant(number: int) -> bool:
    total = 0
    for n in range(1, int((number/2)+1)):
        if number % n == 0:
            total += n
    if total > number:
        return True
    else:
        return False

def generate_abundant_list() -> list:
    abundant_numbers = []
    for number in range(1, 28123):
        if is_abundant(number):
            abundant_numbers.append(number)
    return abundant_numbers

abundant_numbers_list = generate_abundant_list()

print(len(abundant_numbers_list))
```
