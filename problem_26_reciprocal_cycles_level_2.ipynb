```python
def len_repeat(num: int, min_length: int = 0) -> int:
    """
    Finds the length of the repeating cycle of 1 / num.
    """
    seen = {}
    remainder = 1
    position = 0

    while remainder != 0:
        if remainder in seen:
            return position - seen[remainder]
        seen[remainder] = position
        remainder = (remainder * 10) % num
        position += 1
    return 0


def return_cycle(d: int) -> int:
    """
    Goes over the numbers and finds the number with the longest cycle.
    """
    curr_best = 0
    curr_best_num = None

    for num in range(1, d):
        cycle_len = len_repeat(num, curr_best)
        if cycle_len > curr_best:
            curr_best = cycle_len
            curr_best_num = num
    return curr_best_num

print(return_cycle(1000))
```
