```python
def is_prime(num: int) -> bool:
    """
    Returns True if num is prime, otherwise False.
    """
    if num < 2:
        return False

    for i in range(2, int(num ** 0.5) + 1):
        if num % i == 0:
            return False

    return True


def count_streak(a: int, b: int) -> int:
    """
    Counts how many consecutive n-values produce primes
    for n^2 + a*n + b, starting at n = 0.
    """
    n = 0

    while True:
        val = n * n + a * n + b

        if not is_prime(val):
            return n

        n += 1


def main() -> int:
    best_a = 0
    best_b = 0
    best_streak = 0

    for a in range(-999, 1000):
        for b in range(2, 1001):
            if not is_prime(b):
                continue

            streak = count_streak(a, b)

            if streak > best_streak:
                best_streak = streak
                best_a = a
                best_b = b

    return best_a * best_b


print(main())
```
