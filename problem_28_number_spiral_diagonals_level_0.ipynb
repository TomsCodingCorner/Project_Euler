```python
def corner_sum(n: int) -> int:
    """
    Calculate the sum of the four corners of a spiral layer.

    The largest corner is n**2.
    The other corners are found by subtracting n - 1 each time.
    """
    return (n**2 + (n**2 - (n - 1)) + (n**2 - 2 * (n - 1)) + (n**2 - 3 * (n - 1)))


def spiral(size: int) -> int:
    """
    Go layer by layer through the spiral and calculate the diagonal sum.
    """
    total = 1

    for n in range(3, size + 1, 2):
        total += corner_sum(n)
    return total


print(spiral(1001))
```
