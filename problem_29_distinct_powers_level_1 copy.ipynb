```python
def combinations() -> int:
    """
    Returns all the distinct terms
    for 2 <= a <= 100 and 2 <= b <= 100.
    """
    unique_list = []
    for a in range(2, 101):
        for b in range(2, 101):
            combi = a**b
            if combi not in unique_list: 
                unique_list.append(combi)
    return len(unique_list)


print(combinations())
```
