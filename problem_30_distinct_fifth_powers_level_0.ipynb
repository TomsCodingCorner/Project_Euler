```python
def fifth_power() -> int:
    """
    Returns the sum of all numbers that can be written as the sum of fifth powers of their digits.    
    """
    upper = 6 * 9**5
    total_sum = 0
    for number in range(2, upper+1):
        total = 0
        for num in str(number):
            total += int(num) ** 5
        if total == number:
            total_sum += total
    return total_sum

print(fifth_power())
```
