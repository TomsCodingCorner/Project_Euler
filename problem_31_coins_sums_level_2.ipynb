```python
def ways_of_X_pounds(x: int) -> int:
    """
    Returns the amount of different ways to get x pence.    
    """
    coins = [1, 2, 5, 10, 20, 50, 100, 200]

    ways = [0] * (x + 1)
    ways[0] = 1

    for coin in coins:
        for amount in range(coin, x + 1):
            ways[amount] += ways[amount - coin]
    return ways[x]

print(ways_of_X_pounds(200))
```
