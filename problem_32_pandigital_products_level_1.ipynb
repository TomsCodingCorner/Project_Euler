```python
def is_pandigital(a: int, b: int, c: int) -> bool:
    unique = []
    digits = str(a)+str(b)+str(c)

    if len(digits) != 9:
        return False
    
    for i in digits:
        if i == "0":
            return False
        if i in unique:
            return False
        unique.append(i)
    return True

def pandigital() -> int:
    products = set()
    for a in range(1, 100):
        for b in range(1, 10000):
            product = a * b
            if is_pandigital(a, b, product):
                products.add(product)
    return sum(products)


print(pandigital())
```
