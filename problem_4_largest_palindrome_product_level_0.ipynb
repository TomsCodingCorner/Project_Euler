```python
def is_palindroom(num: int) -> bool:
    if str(num)[::-1] == str(num):
        return True
    return False

def largest()-> int:
    largest = 0
    for num1 in range(999, -1, -1):
        for num2 in range(999, -1, -1):
            found = num1*num2
            if is_palindroom(num=found) and found > largest:
                largest = found
    return largest
            
print(largest())
```
