```python
def dif(lim=100) -> int:
    sum_of_squares = 0
    sum_of_numbers = 0
    
    for i in range(1, lim + 1):
        sum_of_squares += i * i
        sum_of_numbers += i
    
    return sum_of_numbers ** 2 - sum_of_squares

print(dif())
```
