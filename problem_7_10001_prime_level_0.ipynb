```python
def num_primes(idx_number = 10001)-> int:
    """
    Based on the code from problem 10, finding idx number prime
    """
    idx = 1 #Waarde 2 doen we we buiten de loop
    num = 3
    while(True):
        for check in range(3, int(num ** 0.5) + 1,2):
            if num % check == 0:
                break
        else:
            idx += 1
            if idx == idx_number:
                return num
        num += 2

print(num_primes())
```
