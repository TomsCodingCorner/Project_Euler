```python
def special_pythagorean_triplet(special_number = 1000)->int:
    """
    The special number is 1000.
    This function implementation of an optimized brute force to get the a + b + c = 1000, pythagororean thriplet.

    """
    for c in range(333, 1000):
        for b in range(c):
            for a in range(b):
                if a**2 + b**2 > c**2:
                    break
                if c + b + 333 < 1000:
                    break
                if a**2 + b**2 == c**2 and (a + b + c) == special_number:
                    return a*b*c

print("The answer is: ", special_pythagorean_triplet())
```
